```python
import matplotlib.pyplot as plt
plt.style.use('smoh')
```

```python
import gapipes as gp

gp.gaia.query("""
select *
from gaiadr2.gaia_source
where
  1=contains(point('', ra, dec),
             circle('', 130.226, 19.665, 1))
  and parallax between 4.613 and 7.312
""").pipe(gp.add_gMag)\
    .plot.scatter('bp_rp', 'gMag').invert_yaxis();

plt.savefig('cmd.png')
```
